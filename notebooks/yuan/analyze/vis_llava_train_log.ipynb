```python
import numpy as np
import json
import os

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def plot_history(model_path):
    with open(os.path.join(model_path, "trainer_state.json"), "r") as fp:
        data = json.load(fp)

    train_history = data["log_history"]

    epochs = [x["epoch"] for x in train_history[:-1]]
    losses = [x["loss"] for x in train_history[:-1]]

    plt.figure()
    plt.plot(epochs, losses)
    plt.show()
```
