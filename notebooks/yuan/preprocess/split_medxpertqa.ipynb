```python
root = "/media/yesindeed/DATADRIVE1/mount/remote_cse/datasets/MedXpertQA"
# annotation_path = os.path.join(root, "MM")

samples = []

with open(os.path.join(root, "MM", "test.jsonl"), "r", encoding="utf-8") as f:
    for line in f:
        record = json.loads(line)
        samples.append(record)

record
```

```python
import random
from typing import List, Any, List, Tuple


def make_five_folds(items: List[Any], seed: int = 42):
    """Shuffle items and split into 5 folds. Returns a list of (train, val) pairs."""
    rng = random.Random(seed)
    items_shuffled = items[:]  # 不修改原列表
    rng.shuffle(items_shuffled)

    n = len(items_shuffled)
    folds = []
    fold_sizes = [(n + i) // 5 for i in range(5)]  # 尽量均匀
    starts = [sum(fold_sizes[:i]) for i in range(5)]
    print(starts)

    for i in range(5):
        s, e = starts[i], starts[i] + fold_sizes[i]
        val = items_shuffled[s:e]
        train = items_shuffled[:s] + items_shuffled[e:]
        folds.append((train, val))
    return folds


def write_list_jsonl(items, file):
    with open(file, "w", encoding="utf-8") as f:
        print(f"Writing {len(items)} objects into {file}")
        for obj in items:
            f.write(json.dumps(obj, ensure_ascii=False) + "\n")


folds = make_five_folds(samples)

for i, fold in enumerate(folds):
    train, val = fold

    write_list_jsonl(train, os.path.join(root, "MM", f"cus_train_fold{i}.jsonl"))
    write_list_jsonl(val, os.path.join(root, "MM", f"cus_test_fold{i}.jsonl"))
```
